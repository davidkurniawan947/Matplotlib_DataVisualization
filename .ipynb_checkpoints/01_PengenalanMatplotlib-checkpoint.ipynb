# **Pengenalan Matplotlib**

- Matplotlib adalah  salah satu module dalam pemrograman Python yang dapat digunakan untuk memvisualisasikan data dalam bentuk grafis
- Dengan Matplotlib, graps akan digambarkan pada suatu **figure**, yang di dalamnya dapat terdiri satu atau lebih axes

# Import modules

```python
%matplotlib inline
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Membuat plotting sederhana

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

fig, ax = plt.subplots() # membuat sebuah figure dan sebuah axes
ax.plot(x,y) # melakukan plotting data pada axes
```

**Alternatif, kita bisa memanfaatkan method plot() pada pyplot untuk melakukan ploting sederhana**

```python
# persiapan sample data
x = [1, 2, 3, 4]
y = [1, 4, 2, 3]

plt.plot(x, y)
```

Komponen dari sebuah Matplotlib **Figure**

Berikut adalah beberapa cara untuk membuat **figure** dan **axes**:

```python
fig = plt.figure() #figure tanpa axes
```

```python
fig, ax = plt.subplots() #sebuah figure dengan sebuah axes
```

```python
fi, axs = plt.subplots(2,3) # sebuah figure dengan 2x3 (baris x kolom) grid axes 
```

# **Ada 2 cara dalam menggunakan Matplotlib**

- OO Style (Objecy Oriented Style)
- pyplot Style

# Mengenal Object Oriented Style untuk melakukan plotting

Secara eksplisit membuat figures beserta axes, dan memanggil methods dari keduanya.

```python
x = np.linspace(0, 2, 100) # (p1, p2, p3) merepresentasikan sebagai (starting_point, n_point, jumlahData_point)
x
```

```python
fig, ax = plt.subplots() # membuat sebuah figure dan sebuah axes

# plotting tiga variant data pada axes
ax.plot(x, x, label='linear')
ax.plot(x, x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x_label') # menyertakan x-label pada axes
ax.set_ylabel('y_label') # menyertakan y-label pada axes
ax.set_title('Simple Plot') # menyertakan title pada axes
ax.legend() # menyertakan legend
```

# **Mengenal pyplot Style untuk melakukan plotting**

**Mengandalkan pyplot untuk membuat dan mengelola figures dan axes, serta menggunakan fungsi pada pyplot untuk melakukan plotting**

```python
x = np.linspace(0, 2, 100) # (p1, p2, p3) merepresentasikan sebagai (starting_point, n_point, jumlahData_point)
x
```

```python
# plotting tiga variant data pada axes
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')

plt.xlabel('x_label') # menyertakan x-label pada axes
plt.ylabel('y_label') # menyertakan y-label pada axes
plt.title('Simple Plot') # menyertakan title pada axes
plt.legend() # menyertakan legend
```
